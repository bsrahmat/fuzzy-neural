```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
```

```python
x = np.arange(0, 255, 1)
a = 100
b = 150
c = 200
y = fuzz.trimf(x, [a, b, c])

plt.plot(y)
plt.xlabel('x')
plt.ylabel('Derajad Keanggotaan Fuzzy')
plt.title('Fungsi Keanggotaan Segitiga')
plt.grid(True)
```

```python
x = np.arange(0, 255, 1)
a = 50
b = 100
c = 150
d = 200
y = fuzz.trapmf(x, [a, b, c, d])

plt.plot(y)
plt.xlabel('x')
plt.ylabel('Derajad Keanggotaan Fuzzy')
plt.title('Fungsi Keanggotaan Trapesium')
plt.grid(True)
```

```python
x = np.arange(0, 255, 1)
a = 125
b = 50
y = fuzz.gaussmf(x, a, b)

plt.plot(y)
plt.xlabel('x')
plt.ylabel('Derajad Keanggotaan Fuzzy')
plt.title('Fungsi Keanggotaan Gaussian')
plt.grid(True)
```
