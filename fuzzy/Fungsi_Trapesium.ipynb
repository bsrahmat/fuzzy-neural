```python
import numpy as np
import matplotlib.pyplot as plt
import skfuzzy as fuzz
x = np.arange(0, 255, 1)
a = 50
b = 100
c = 150
d = 200
y = fuzz.trapmf(x, [a, b, c, d])
plt.plot(y)
plt.xlabel('x')
plt.ylabel('Derajad Keanggotaan Fuzzy')
plt.title('Fungsi Keanggotaan Trapesium')
plt.grid(True)
```
